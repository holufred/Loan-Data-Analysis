## Investigation Overview


> Loans for Debt Consolidation, Business and Baby & Adoption gets the approval for higher amounts. Out of these three categories, Debt Consolidation is the money maker as it's the most popular and is projected to bring in more returns than other categories. Health workers, salespeople and computer programmers are the top people who love consolidating their debt. While healthworkers fetches the most returns, computer programmers are less likey to default. The analysis also reveal that approval for 3-year loans for this category should be dialled down. Focus should instead be on the 1-year loan as there is lesser chance of default


## Dataset Overview

> This Loan Data from Prosper (a peer-to-peer lending platform) contains 113,937 loan entries with 81 variables covering a range of loan terms, borrower demographics, and credit profiles. Loan entries listed before July 2009 were dropped and at the end of the wrangling processes, there were 76168 entries with 19 variables to explore

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
prosper =pd.read_csv('prosperLoanData.csv')
#select the features of interest
select_features =['ListingKey', 'LoanOriginationDate', 'LoanStatus',  
                  'Term', "ListingCreationDate", 'IncomeRange', 'IncomeVerifiable',
                  'EstimatedReturn', 'ListingCategory (numeric)',  'ProsperRating (Alpha)', 
                  'BorrowerState', 'Occupation', 'EmploymentStatus', 'EmploymentStatusDuration', 
                  'IsBorrowerHomeowner', 'OpenCreditLines', 'OpenRevolvingAccounts', 'InquiriesLast6Months', 
                  'CurrentDelinquencies',  'LoanOriginalAmount', 'DebtToIncomeRatio']
prosper_new=prosper[select_features]

#filter out entries earlier than July 2009 when most of the features takes effect to avoid null values on important variables
prosper_new=prosper_new.loc[pd.to_datetime(prosper_new.ListingCreationDate) >= '2009-07-31']

#rename columns
prosper_new.rename(columns={'ProsperRating (Alpha)': 'ProsperRating', "ListingCategory (numeric)": "ListingCategory"}, inplace=True)


#filter off entries where the income is not verifiable
prosper_new= prosper_new[prosper_new.IncomeVerifiable ==True]

#drop superfluous columns
prosper_new.drop(["ListingCreationDate",'IncomeVerifiable'], axis=1, inplace=True)
prosper_new =prosper_new.loc[~((prosper_new.EmploymentStatusDuration.isnull()) |(prosper_new.DebtToIncomeRatio.isnull()))]

#Drop the entries with null values in the Occupation column
prosper_new=prosper_new.loc[~(prosper_new.Occupation.isnull())]

#Recategorize
prosper_new.EmploymentStatus=prosper_new.EmploymentStatus.replace({'Full-time': 'Employed', 'Part-time': 
                                                                   'Employed', 'Self-employed': 'Employed', 
                                                                  'Retired': 'Not employed'})
#Change employment status fo some entries
prosper_new.loc[(~prosper_new.Occupation.isin(['Other','Others'])), 'EmploymentStatus'] = 'Employed'

#Convert ProsperRating to a category
rating_cat =['HR','E','D','C','B','A','AA']
credit_cat =pd.api.types.CategoricalDtype(categories=rating_cat, ordered=True)
prosper_new.ProsperRating=prosper_new.ProsperRating.astype(credit_cat)

#Recategorize
prosper_new.LoanStatus.replace({'Past Due (>120 days)': 'Not Completed', 
                                'Past Due (1-15 days)': 'Not Completed',
                                'Past Due (16-30 days)': 'Not Completed',
                                'Past Due (31-60 days)': 'Not Completed',
                                'Past Due (61-90 days)': 'Not Completed',
                                'Past Due (91-120 days)': 'Not Completed',
                                'Chargedoff': 'Not Completed',
                                'Defaulted': 'Not Completed',
                                'FinalPaymentInProgress': 'Completed'
                               }, inplace=True)

#Combine some subcategories
prosper_new.Occupation= (prosper_new.Occupation
                             .str.strip()
                             .str.replace(r'(Sales)\s+-\s+\w+', 'Salesperson')
                             .str.replace(r'(Engineer)\s+-\s+\w+', 'Engineer')
                             .str.replace(r'(Teacher)(\'s)?(\s+)?(\w+)?', 'Teacher') #including Teacher's Aide
                             .str.replace(r'(Pilot)\s+-\s+\w+/\w+', 'Pilot')
                             .str.replace(r'(Tradesman)\s+-\s+\w+', 'Tradesman')
                             #Create a 'Health Worker' occupation
                             .str.replace(r'(Nurse)(\'s)?(\s+)?\(?\w+\)?', 'Health Worker') #including Nurse's Aide
                             .str.replace('Medical Technician', 'Health Worker')
                             .str.replace('Pharmacist', 'Health Worker')
                             .str.replace('Doctor', 'Health Worker')
                             .str.replace('Dentist', 'Health Worker')
                             .str.replace(r'(Student)\s+-\s+\w+(\s+\w+)?(\s+\w+)?', 'Student')
                         
                             .str.replace('Others','Other')
                        )
```

## Debt Consolidation Loans

> Debt consolidation loans account for about 70% of projected returns and 64% of all loan entries. From the chart, it appears that folks want to put their house in order and consolidate all their debt in anticipation of a new year as there's a surge from October through January 

```python
months=['January', 'February', 'March', 'April','May', 'June', 
        'July', 'August', 'September', 'October', 'November', 'December']

month_loan =(prosper_new
        #select the Debt Consolidation Category
     .loc[prosper_new.ListingCategory ==1]
        #Extract the month from the date the loan was disbursed
     .assign(LoanMonth= pd.to_datetime(prosper_new.LoanOriginationDate).dt.month_name())
        #groupby by the date and reindex to ensure the months are in the right order
     .pipe(lambda df: df.groupby(['LoanMonth','ListingCategory'])['ListingKey'].count().unstack().reindex(months[::1]))
     )


ax = month_loan.plot(kind='line', color='#ff6361', grid=True, figsize=(16,8))

#Ensure that the months are showing on the x-axis
ax.set_xticks(range(len(months)))
ax.set_xticklabels(months, rotation=45, ha='right')

#Set a style for the labels
font_style = {'family': 'DejaVu Sans', 'color': '#333333', 'size': 14}

# Set the font style for the title
title_style = {'family': 'DejaVu Sans', 'color': '#333333', 'size': 20}

ax.set_title('Monthly Comparison of Debt Consolidation Loans Disbursal',  fontdict=title_style)

#remove the superfluous legend to prevent chart junk
ax.legend().remove()

plt.ylabel('Number of Loans Disbursed', fontdict=font_style)
plt.xlabel('Month', fontdict=font_style);

```

## Occupation Comparison

> While healthworkers, salespeople and computer programmers are estimated to bring the largest returns to the firm with debt consolidation loans, computer programmers are less likely to default or miss a payment followed by salespeople (who are in the sweet middle) and then healthworkers.

```python
#Create font for labels
font = {'family': 'serif',
        'color':  '#d45087',
        'weight': 'normal',
        'size': 16,
        }

#Create a new 'LoanReturn' feature by multiplying the estimated return percentage with the original loan amount 
prosper_new['LoanReturn']= prosper_new.LoanOriginalAmount.mul(prosper_new.EstimatedReturn)

# Plot 1
top_debt_payment = (prosper_new
                     .loc[prosper_new.ListingCategory == 1] #which is the dummy for Debt Consolidation
                     #filter off "Other", "Professional", and "Executive"
                     .loc[~(prosper_new.Occupation.isin(['Other', 'Professional', 'Executive']))]
                     #filter of current loans
                    .loc[~(prosper_new.LoanStatus =='Current')]
                     #group by Occupation and select the first 
                     .pipe(lambda df: df.groupby(['Occupation', 'LoanStatus'])['ListingKey'].count().transform(lambda x: x/x.sum()*100).unstack().sort_values(by='Completed',ascending=False))[:5]
          )

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18,12))
fig.suptitle('Occupation Analysis for Debt Consolidation Loans', y=1.0, fontsize=18, fontweight='bold')
# Plot 1 - Top 5 Occupations that Apply for Debt Consolidation Loans
top_debt_payment.plot(kind='bar',color =['#665191','#d45087'],title='Top 5 Occupations that Apply for Debt Consolidation Loans', ax=ax1)
ax1.set_xlabel('Occupation', fontdict=font)
ax1.set_ylabel('Percentage of Loan Entries', fontdict=font)
# Move the legend outside of the chart to make the chart more smoother
ax1.legend(loc='best')
# Add values on top of the bars
for p in ax1.patches:
    ax1.annotate('{:.2f}%'.format(p.get_height()), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', xytext=(0, 5), 
                textcoords='offset points')

# Plot 2
total_debt = (prosper_new
                     .loc[prosper_new.ListingCategory == 1] #which is the dummy for Debt Consolidation
                     #filter off "Other", "Professional", and "Executive"
                     .loc[~(prosper_new.Occupation.isin(['Other', 'Professional', 'Executive']))]
                     .loc[~(prosper_new.LoanStatus == 'Not Completed')] #filter off defaulted loans
                     #group by Occupation and select the first 
                     .pipe(lambda df: df.groupby('Occupation')['LoanReturn'].sum().sort_values(ascending=False))[:5]
          )


total_debt.plot(kind='bar',color =['#665191'],title='Top 5 Occupations that Fetches the Largest Returns', ax=ax2)
ax2.set_xlabel('Occupation',  fontdict=font)
ax2.set_ylabel('Total Returns Estimated', fontdict=font)
# Add values on top of the bars
for p in ax2.patches:
    ax2.annotate('${:,.0f}'.format(p.get_height()), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', xytext=(0, 5), 
                textcoords='offset points')

plt.tight_layout(pad=3)
```

## Best Loan Term
It appears that 3-year loans account for majority of debt consolidation loans defaulted or not paid back completely while  1-year loans are the least likely to be defaulted. The 1-year Debt Consolidation loans should be marketed more to folks particularly healthworkers, salespeople and computer programmers from the end of September each year.

```python
top_debt_term =(prosper_new
                     .loc[prosper_new.ListingCategory == 1] #which is the dummy for Debt Consolidation
                     #filter off "Other", "Professional", and "Executive"
                     .loc[(prosper_new.Occupation .isin(['Health Worker', 'Salesperson', 'Computer Programmer']))]
                     #filter of current loans
                    .loc[(prosper_new.LoanStatus =='Not Completed')]
                     #group by Occupation and select the first 
                     .pipe(lambda df: df.groupby(['Occupation', 'Term'])['ListingKey'].count().groupby(level=0).transform(lambda x: x/x.sum()*100).sort_values().unstack())[:5]
          )

ax = top_debt_term.plot(kind='bar',color =['#ffa600','#d45087','#003f5c'],title='Term times for Non Completed Debt Consolidation Loans',rot=0,figsize=(14,6))
plt.xlabel('Occupation')
plt.ylabel('Percentage of Loan Entries')

# Move the legend outside of the chart to make the chart more smoother
ax.legend(title= 'Loan Term',loc='best')

# Add values on top of the bars
for p in ax.patches:
    ax.annotate('{:.2f}%'.format(p.get_height()), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', xytext=(0, 5), 
                textcoords='offset points')
```
