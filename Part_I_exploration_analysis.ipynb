## Preliminary Wrangling


```python
#select the features of interest
select_features =['ListingKey', 'LoanOriginationDate', 'LoanStatus',  
                  'Term', "ListingCreationDate", 'IncomeRange', 'IncomeVerifiable',
                  'EstimatedReturn', 'ListingCategory (numeric)',  'ProsperRating (Alpha)', 
                  'BorrowerState', 'Occupation', 'EmploymentStatus', 'EmploymentStatusDuration', 
                  'IsBorrowerHomeowner', 'OpenCreditLines', 'OpenRevolvingAccounts', 'InquiriesLast6Months', 
                  'CurrentDelinquencies',  'LoanOriginalAmount', 'DebtToIncomeRatio']
prosper_new=prosper[select_features]

# prosper_new.ListingCreationDate
#filter out entries earlier than July 2009 when most of the features takes effect to avoid null values on important variables
prosper_new=prosper_new.loc[pd.to_datetime(prosper_new.ListingCreationDate) >= '2009-07-31']

#rename columns
prosper_new.rename(columns={'ProsperRating (Alpha)': 'ProsperRating', "ListingCategory (numeric)": "ListingCategory"}, inplace=True)


#filter off entries where the income is not verifiable
prosper_new= prosper_new[prosper_new.IncomeVerifiable ==True]

#drop superfluous columns
prosper_new.drop(["ListingCreationDate",'IncomeVerifiable'], axis=1, inplace=True)

prosper_new.head()
```

```python
#Check dtype
prosper_new.dtypes
```

```python
#Check for null values
prosper_new.isnull().sum()
```

#### Drop the entries containing null values for EmploymentStatusDuration and DebtToIncomeRatio since they are small. The null values in the Occupation variable would be  further examined

```python
prosper_new =prosper_new.loc[~((prosper_new.EmploymentStatusDuration.isnull()) |(prosper_new.DebtToIncomeRatio.isnull()))]
```

```python
#Examine the 1327 NaN values in the Occupation column
(prosper_new
     .loc[prosper_new.Occupation.isnull()]
     .loc[:,['Occupation', 'EmploymentStatus','EmploymentStatusDuration','IncomeRange']]
     .sort_values(by='EmploymentStatus')
     .head(10)
)
```

#### It seems only one person is employed out of the people with null values in the occupation column.

```python
#Drop the entries with null values in the Occupation column
prosper_new=prosper_new.loc[~(prosper_new.Occupation.isnull())]
```

```python
#Check out the Employement Status and clean if neccessary
prosper_new.EmploymentStatus.value_counts()
```


>1. Group full-time, part-time, and self-employed as employed to reduce the categories. 
>2. Group Retired as 'Not employed'
>3. If the corresponding Occupation value is not 'Other' or 'Others, change the EmploymentStatus to 'Employed' as this means the person has a job

```python
prosper_new.EmploymentStatus=prosper_new.EmploymentStatus.replace({'Full-time': 'Employed', 'Part-time': 
                                                                   'Employed', 'Self-employed': 'Employed', 
                                                                  'Retired': 'Not employed'})

prosper_new.loc[(~prosper_new.Occupation.isin(['Other','Others'])), 'EmploymentStatus'] = 'Employed'
#Test
prosper_new.EmploymentStatus.value_counts(dropna=False)
```

```python
#Check out the new structure
prosper_new.shape
```

```python
#Check out the credit rating (ProsperRating) variable
prosper_new.ProsperRating.value_counts(dropna=False)
```

```python
#Convert ProsperRating to a category
rating_cat =['HR','E','D','C','B','A','AA']
credit_cat =pd.api.types.CategoricalDtype(categories=rating_cat, ordered=True)
prosper_new.ProsperRating=prosper_new.ProsperRating.astype(credit_cat)
prosper_new.ProsperRating.dtypes
```

```python
#Examine LoanStatus variable
prosper_new.LoanStatus.value_counts(dropna=False)
```

#### Reorder the variable as follows:
1. Change 'Defaulted', 'Chargedoff' and all 'Past Due' buckets to  'Not Completed' as they have a higher chance of not completing
2. Merge 'FinalPaymentInProgress with 'Completed' as they are about to make the final payment to complete the loan
3. Leave'Current' Status as it is as they have equal chance of defaulting or completing all  loan payment


```python
prosper_new.LoanStatus.replace({'Past Due (>120 days)': 'Not Completed', 
                                'Past Due (1-15 days)': 'Not Completed',
                                'Past Due (16-30 days)': 'Not Completed',
                                'Past Due (31-60 days)': 'Not Completed',
                                'Past Due (61-90 days)': 'Not Completed',
                                'Past Due (91-120 days)': 'Not Completed',
                                'Chargedoff': 'Not Completed',
                                'Defaulted': 'Not Completed',
                                'FinalPaymentInProgress': 'Completed'
                               }, inplace=True)
#Test
prosper_new.LoanStatus.value_counts()
```

```python
#Group by income range
income_plot =prosper_new.groupby('IncomeRange')['ListingKey'].size()

#Create a function for plotting
def plotter(frame, color_code):
    #Get the total entries
    total=frame.size
    
    #plot
    if frame.index.size < 7:
        ax =frame.plot(kind='bar',figsize= (11,8),color =color_code)
        for i, v in enumerate(frame.transform(lambda x: (x/x.sum())*100)):
            ax.text(i, (v*total*0.01)+1000, f'{str(round(v,2))}%', ha='center', fontweight='bold')
    else:
        ax =frame.plot(kind='barh',figsize= (11,8),color =color_code)
        for i, v in enumerate(frame.transform(lambda x: (x/x.sum())*100)):
            ax.text((v*total*0.01)+1000, i, f'{str(round(v,2))}%', ha='center', fontweight='bold')
    
    return ax
    
plotter(income_plot, '#58508d')
plt.xlabel('Income Range')
plt.ylabel('Number of Loan Clients')
plt.suptitle('Distribution of Loan Clients Per Income Range')
plt.tight_layout(pad=3);
```

#### Observation: Seems people in the $50,000-74,999 income bracket takes out the most loans

#### Question: Still on Debt Consolidation, which top occupations apply for debt consolidation and what percentage eventually complete payment?

```python
top_debt_payment =(prosper_new
                     .loc[prosper_new.ListingCategory == 1] #which is the dummy for Debt Consolidation
                     #filter off "Other", "Professional", and "Executive"
                     .loc[~(prosper_new.Occupation .isin(['Other', 'Professional', 'Executive']))]
                     #filter of current loans
                    .loc[~(prosper_new.LoanStatus =='Current')]
                     #group by Occupation and select the first 
                     .pipe(lambda df: df.groupby(['Occupation', 'LoanStatus'])['ListingKey'].count().transform(lambda x: x/x.sum()*100).unstack().sort_values(by='Completed',ascending=False))[:5]
          )
top_debt_payment
ax = top_debt_payment.plot(kind='bar',color =['#665191','#d45087'],title='Top 5 Occupations that Apply for Debt Consolidation Loans',rot=0,figsize=(14,6))
plt.xlabel('Occupation')
plt.ylabel('Percentage of Loan Entries')

# Move the legend outside of the chart to make the chart more smoother
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

# Add values on top of the bars
for p in ax.patches:
    ax.annotate('{:.2f}%'.format(p.get_height()), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', xytext=(0, 5), 
                textcoords='offset points')
```

#### Observations: Computer programmers  and salespeople seems to love this type of loans.
